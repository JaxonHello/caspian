```python
def readTAD(tadfile):
    #tads = "/home/ghaiyan/project/CASPIAN/evaluate_TADS/GM12878/chr19_5kb/TAD/{}.txt".format(tadsname)
    f = open(tadfile)
    line=f.readline()
    start=[]
    end=[]
    while line:
        line = line.split()
        start1 = int(float(line[0]))
        end1 = int(float(line[2]))
        start.append(start1)
        end.append(end1)
        line=f.readline()
    f.close()
    return start, end
```
